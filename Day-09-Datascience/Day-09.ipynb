# 1️⃣ Introduction to Matplotlib

Matplotlib is one of the most widely used Python libraries for **data visualization**.  
It allows data scientists and analysts to create charts such as line plots, bar charts, histograms, and scatter plots.

Matplotlib is commonly used in:
- Data analysis
- Machine learning projects
- Exploratory data analysis
- Scientific research

The most commonly used module is:

`matplotlib.pyplot`

---

# 2️⃣ What is Data Visualization

Data visualization is the process of **representing data using charts and graphs** so that patterns, trends, and relationships become easier to understand.

Examples of visualizations:
- Line charts for trends
- Bar charts for category comparison
- Histograms for distributions
- Scatter plots for relationships between variables

Good visualization helps transform **raw numbers into meaningful insights**.

# 4️⃣ Components of a Figure

A Matplotlib chart contains several components.

Main components include:

**Figure**  
The entire canvas where the chart is drawn.

**Axes**  
The area where the data is plotted.

**Title**  
The heading of the chart.

**X-axis Label**  
Description of the horizontal axis.

**Y-axis Label**  
Description of the vertical axis.

**Legend**  
Explains multiple lines or categories in a chart.

---

# 5️⃣ Plot Function

The `plot()` function is the most basic plotting function in Matplotlib.

It is mainly used to create **line plots**.

Example syntax:

```python
plt.plot(x, y)
```

After plotting, the chart is displayed using:

```python
plt.show()
```

---

# 6️⃣ Univariate Analysis

Univariate analysis involves analyzing **a single variable**.

Example variables:
- Age
- Salary
- Height
- Exam scores

Common plots used for univariate analysis include:
- Histogram
- Boxplot
- Pie chart

---

# 7️⃣ Line Plot

A line plot is used to visualize **trends over time**.

Examples:
- Stock price trends
- Monthly sales growth
- Temperature changes

---

# 8️⃣ Histogram

A histogram shows the **distribution of numerical data**.

It groups values into ranges called **bins**.

Example uses:
- Age distribution
- Salary ranges
- Exam scores

---

# 9️⃣ Boxplot

A boxplot summarizes the distribution of data.

It shows:
- Minimum value
- First quartile (Q1)
- Median
- Third quartile (Q3)
- Maximum value

Boxplots help detect **outliers**.

---

# 🔟 Pie Chart

A pie chart represents **percentage contribution** of categories.

Examples:
- Budget allocation
- Market share
- Department distribution

---

# 1️⃣1️⃣ Countplot

A countplot shows **frequency of categorical variables**.

Example:
- Number of employees per department
- Number of students per major

---

# 1️⃣2️⃣ Bivariate Analysis

Bivariate analysis studies the relationship between **two variables**.

Examples:
- Experience vs Salary
- Study Hours vs Exam Score

---

# 1️⃣3️⃣ Scatter Plot

Scatter plots visualize the relationship between **two numerical variables**.

Each point represents one observation.

Useful for detecting:
- Correlation
- Patterns
- Trends

---

# 1️⃣4️⃣ Bar Chart

Bar charts compare **categorical data**.

Examples:
- Employees per department
- Sales per product
- Population per country

---

# 1️⃣5️⃣ Bivariate Analysis (Numerical vs Categorical)

Sometimes one variable is **categorical** and the other is **numerical**.

Example:

Department → Category  
Salary → Numeric  

Visualizations used:
- Bar chart
- Boxplot

---

# 1️⃣6️⃣ Multivariate Analysis

Multivariate analysis examines **more than two variables simultaneously**.

Example variables:
- Experience
- Salary
- Age
- Department

This helps uncover **complex relationships in datasets**.

---

# 1️⃣7️⃣ Bubble Plot

A bubble plot extends a scatter plot by adding **bubble size as a third variable**.

Example:

X-axis → Experience  
Y-axis → Salary  
Bubble size → Age

---

# 1️⃣8️⃣ Object-Oriented API

Matplotlib also supports an **Object-Oriented approach**.

Example:

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

This approach provides **better control over visualizations**.

---

# 1️⃣9️⃣ Multiple Plots (Subplots)

Subplots allow multiple charts in a **single figure**.

Example dashboard:
- Histogram of ages
- Scatter plot of experience vs salary
- Bar chart of department employees

---

# 2️⃣0️⃣ 3D Plot

Matplotlib supports **3D visualizations**.

Examples:
- 3D scatter plots
- 3D surface plots
- 3D line plots

These plots visualize relationships across **three dimensions**.

# **Bivariate - mean two variable (Numerical - Categorical)**


````python
# creating a function which shows the salary of the persons working in Hr

salary_hr = salary_new[salary_new['departments'] == 'HR']
salary_hr
````

````python
# creating a function which shows the salary of the persons working in Finance

salary_finance = salary_new[salary_new['departments'] == 'Finance']
salary_finance
````

````python
# creating a function which shows the salary of the persons working in IT

salary_IT = salary_new[salary_new['departments'] == 'IT']
salary_IT
````

````python
# creating a function which shows the salary of the persons working in Marketing

salary_marketing = salary_new[salary_new['departments'] == 'Marketing']
salary_marketing
````

# Box Plot

````python
plt.boxplot([salary_IT['salary'], salary_finance['salary'], salary_hr['salary'], salary_marketing['salary']], tick_labels=['IT', 'Finance', 'HR', 'Marketing'], label= ['HR','IT', 'MRKT', 'FIN'])
plt.legend()
plt.show()
````

# Pie Chart

````python
# first we will gropu the salary by departments

salary_group = salary_new.groupby('departments')['salary'].sum()
salary_group
````

````python
# now plot it in pie cahrt

plt.pie(salary_group,labels= salary_group.index, autopct='%1.2f%%')
plt.axis('equal')
plt.show()
````

# Bar plot

````python
# using the averrage salry for the charts

avg_salary_it = sum(salary_IT['salary'])/len(salary_IT['salary'])
avg_salary_finance = sum(salary_finance['salary'])/len(salary_finance['salary'])
avg_salary_hr = sum(salary_hr['salary'])/len(salary_hr['salary'])
avg_salary_marketing = sum(salary_marketing['salary'])/len(salary_marketing['salary'])
````

````python
# Now presenting it on bar chart

plt.bar(['IT', 'Finance', 'HR', 'Marketing'], [avg_salary_it, avg_salary_finance, avg_salary_marketing, avg_salary_hr], color = ['red', 'yellow', 'green', 'orange'])
plt.title("Average Salary")
plt.xlabel("Departments")
plt.ylabel("Salary")
plt.grid()
plt.show()
````

# **Multivariate analysis** : 3 Numerical Columns

````python
# first adding new numerical column

salary_new['experience'] = [0.5, 1, 1.2, 1.5, 2, 2.3, 2.8, 3, 3.2, 3.5,
              4, 4.5, 5, 5.2, 5.8, 6, 6.5, 7, 7.3, 8,
              8.5, 9, 9.5, 10, 11, 12, 13, 14, 15, 18]
````

````python
salary_new.head(5)
````

# Bubble Plot

````python
# Bubble plot
plt.scatter(salary_new['age'], salary_new['salary'], s = salary_new['experience']*50, color = 'skyblue', edgecolors='Black')
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary vs Experience")

plt.show()
````

````python
# when we have 2 numberical column and one categorical column

plt.scatter(salary_new['age'], salary_new['salary'], c = salary_new['departments'].map({'IT': 'red', 'HR': 'blue', 'Finance': 'green', 'Marketing': 'yellow'}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary vs Department")

plt.show()
````

# Object Oriented API


````python
 # Object oriented API
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

# Line plot
axs[0].plot(sort_age['age'], salary_new['salary'], color = "purple", marker = "o")
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Salary')

# histogram

axs[1].hist(salary_new['salary'], bins=10, color='green')
axs[1].set_xlabel('Salary')
axs[1].set_ylabel('Frequency')




# boxplot

axs[2].boxplot(salary_new['salary'])
axs[2].set_title('Boxplot')
axs[2].set_xlabel('Salary')

plt.savefig("mulitplots.png")
plt.tight_layout()
plt.show()


````

````python
data = {
    'Year': [2020, 2021, 2022, 2023],
    'Sales': [100, 150, 200, 250],
    'Profit': [20, 30, 40, 50],
    'Expenses': [80, 120, 160, 200]
}
````

````python
new_data = pd.DataFrame(data)
new_data
````

````python
# using multiple plots

plt.plot(new_data['Year'], new_data['Sales'], label = "Sales" )
plt.plot(new_data['Year'], new_data['Profit'], label = "Profit" )
plt.plot(new_data['Year'], new_data['Expenses'], label = "Expenses" )
plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Sales, Profit, and Expenses Over Time")
plt.legend()
plt.show()
````

````python
plt.plot(new_data['Year'], new_data['Sales'], label = "Sales" )
````

````python
# Creating the 3d Plot

ax = plt.axes(projection='3d')

ax.scatter3D(new_data['Year'], new_data['Sales'], new_data['Profit'])
ax.set_xlabel('Year')
ax.set_ylabel('Sales')
ax.set_zlabel('Profit')
plt.show()
````
