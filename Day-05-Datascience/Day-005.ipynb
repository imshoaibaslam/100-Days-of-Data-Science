## Step 1 вЂ” Understanding Data Types

### What is a `dtype`?

Every column in a DataFrame has a **data type** вЂ” it tells pandas how to store and process that column internally.

| dtype | Meaning |
|-------|---------|
| `float64` | Decimal numbers |
| `int64` | Whole numbers |
| `object` | Text / strings |
| `bool` | True / False |

вљ пёЏ **Important:** String columns do not get a `string` type in pandas.
They are stored as `object` вЂ” keep this in mind when inspecting your data.

#### `dtype` вЂ” Check a Single Column

#### Index `dtype`

## Step 2 вЂ” Converting Data Types with `astype()`

### What is `astype()`?

Sometimes the data type pandas assigned is not the one you need for your analysis.
`astype()` lets you **convert a column from one type to another**.

Example use case:
- `points` is stored as `int64` but your model expects `float64`
- A `0/1` column stored as `int64` should be `bool`

вљ пёЏ The conversion only makes sense if the values are compatible.
Converting a text column to `int64` will throw an error.

````python
# Convert points from int64 to float64
reviews.points.astype('float64')
````

**What just happened?**

- `points` was originally stored as whole numbers (`int64`)
- `astype('float64')` converted every value to a decimal
- 87 в†’ 87.0 вЂ” same value, different storage type
- Original column is untouched вЂ” `astype()` returns a new Series

## Step 3 вЂ” Finding Missing Values

### What is `NaN`?

Missing values in pandas are stored as `NaN` вЂ” short for **Not a Number**.

```
Real value  в†’  87, 15.0, "Italy"
Missing     в†’  NaN
```

вљ пёЏ NaN values are always `float64` dtype вЂ” even if the column contains integers.
This is why a column of whole numbers sometimes shows as `float64` вЂ” it has at least one missing value.

### `pd.isnull()` vs `pd.notnull()`

| Function | Returns |
|----------|--------|
| `pd.isnull()` | `True` where value is NaN |
| `pd.notnull()` | `True` where value is NOT NaN |

#### Find All Rows Where Country is Missing

````python
# Show all rows where country has no value
reviews[pd.isnull(reviews.country)]
````

#### Count Missing Values Per Column

````python
# How many missing values does each column have?
reviews.isnull().sum()
````

---

## Step 4 вЂ” Filling Missing Values with `fillna()`

### What is `fillna()`?

Instead of dropping rows with missing values, we can **fill them** with something meaningful.

Two common strategies:
- Fill with a **constant** вЂ” replace NaN with a default label like `"Unknown"`
- Fill with **backfill** вЂ” use the next non-null value that appears after the missing one

#### Fill Missing Values with a Constant

````python
# Replace every NaN in region_2 with the string "Unknown"
reviews.region_2.fillna("Unknown")
````

**What just happened?**

- Every row where `region_2` was `NaN` now shows `"Unknown"`
- Rows that already had a value are untouched
- Original data is not modified вЂ” `fillna()` returns a new Series

#### Fill Missing Values with the Column Mean

````python
# Replace missing prices with the average price
reviews.price.fillna(reviews.price.mean())
````

## Step 5 вЂ” Replacing Values with `replace()`

### What is `replace()`?

`replace()` is used when you want to **swap out a specific existing value** вЂ” not just fill NaN.

Real-world use case:
- A reviewer changed their Twitter handle
- A category label was renamed
- A sentinel value like `"N/A"` or `"Invalid"` needs to be replaced

````python
# Reviewer Kerin O'Keefe changed her handle from @kerinokeefe to @kerino
# Update the entire column to reflect this
reviews.taster_twitter_handle.replace("@kerinokeefe", "@kerino")
````

**What just happened?**

- Every occurrence of `@kerinokeefe` across all 129,971 rows was replaced with `@kerino`
- All other values remained unchanged
- Original column untouched вЂ” returns a new Series

#### `fillna()` vs `replace()` вЂ” Key Difference

| | `fillna()` | `replace()` |
|-|-----------|------------|
| Works on | NaN values only | Any existing value |
| Use when | Values are missing | Values exist but are wrong |

---

## рџ’Ў When to Use What

| Situation | Use |
|-----------|-----|
| Check type of one column | `df['col'].dtype` |
| Check types of all columns | `df.dtypes` |
| Convert column to different type | `astype('float64')` |
| Find rows with missing values | `pd.isnull(df['col'])` |
| Count missing values per column | `df.isnull().sum()` |
| Fill NaN with a default value | `fillna('Unknown')` |
| Fill NaN with column average | `fillna(df['col'].mean())` |
| Replace a specific existing value | `replace('old', 'new')` |

---

## рџЊЌ Real World Usage

**`dtypes` + `astype()`**
The first check after loading any new dataset. Wrong data types cause silent bugs in machine learning pipelines вЂ” a price column stored as `object` instead of `float64` will break every numeric operation downstream.

**`isnull().sum()`**
Standard step in every Exploratory Data Analysis. Tells you exactly which columns have missing data and how much вЂ” critical before any modelling or aggregation.

**`fillna()`**
Used in data cleaning pipelines before training ML models. Models cannot handle NaN values вЂ” filling them with the mean, median, or a placeholder is a standard preprocessing step.

**`replace()`**
Used when datasets contain sentinel values like `"N/A"`, `"Unknown"`, `"Invalid"` вЂ” or when category labels change over time and historical data needs updating.

---

## рџ”‘ Key Takeaways

- Every column has a `dtype` вЂ” always check it after loading a new dataset
- String columns are stored as `object` вЂ” not `string`
- Missing values are stored as `NaN` and are always `float64` dtype
- `astype()` converts types but does not modify the original column
- `fillna()` fills NaN values вЂ” `replace()` swaps any existing value
- Neither `fillna()` nor `replace()` modifies original data вЂ” they return new copies
- Always run `df.isnull().sum()` as part of your first look at any new dataset
