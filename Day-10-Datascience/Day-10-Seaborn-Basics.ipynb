## Step 1 — What is Seaborn?

Seaborn is a Python data visualization library **built on top of Matplotlib**.

It provides:
- A higher-level interface for drawing statistical graphics
- Beautiful default themes and color palettes
- Built-in support for Pandas DataFrames
- Less code to produce better-looking charts

> **Key difference:** Matplotlib gives you control. Seaborn gives you speed.

## Step 2 — Seaborn vs Matplotlib

| Feature | Matplotlib | Seaborn |
|---------|------------|--------|
| Syntax complexity | High | Low |
| Default aesthetics | Basic | Beautiful |
| Statistical plots | Manual | Built-in |
| DataFrame support | Limited | Native |
| Customization | Full control | Moderate |
| Learning curve | Steep | Gentle |

## Step 3 — Setting Up Seaborn

```python
# Install seaborn (if not already installed)
# !pip install seaborn

# Importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print("Libraries imported successfully!")
print(f"Seaborn version: {sns.__version__}")
```

```python
# Load the Winemag dataset
df = pd.read_csv(r"C:\Users\aslam\Downloads\winemag-data-130k-v2.csv\winemag-data-130k-v2.csv", index_col=0)
df.head()
```

```python
# Quick overview of the dataset
print("Shape:", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())
```

---
## Step 4 — Univariate Analysis

Univariate analysis = analyzing **one variable** at a time.

```python
# 4.1 — Histogram with KDE (Distribution Plot)
plt.figure(figsize=(10, 5))
sns.histplot(df['points'], kde=True, color='purple', bins=20)
plt.title("Distribution of Wine Points")
plt.xlabel("Points")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4.2 — Boxplot (Outlier Detection)
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['price'], color='skyblue')
plt.title("Boxplot of Wine Price")
plt.ylabel("Price ($)")
plt.show()
```

```python
# 4.3 — KDE Plot (Smooth Distribution)
plt.figure(figsize=(10, 5))
sns.kdeplot(df['points'], fill=True, color='green')
plt.title("KDE Plot of Wine Points")
plt.xlabel("Points")
plt.show()
```

---
## Step 5 — Bivariate Analysis

Bivariate analysis = analyzing **two variables** together.

```python
# 5.1 — Scatter Plot (Points vs Price)
df_clean = df[['points', 'price']].dropna()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='points', y='price', alpha=0.4, color='coral')
plt.title("Wine Points vs Price")
plt.xlabel("Points")
plt.ylabel("Price ($)")
plt.show()
```

```python
# 5.2 — Line Plot with Confidence Interval
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_clean, x='points', y='price', color='navy')
plt.title("Avg Price by Points Score")
plt.xlabel("Points")
plt.ylabel("Average Price ($)")
plt.show()
```

```python
# 5.3 — Regression Plot (with trend line)
plt.figure(figsize=(10, 6))
sns.regplot(data=df_clean, x='points', y='price', scatter_kws={'alpha': 0.3}, color='teal')
plt.title("Points vs Price with Regression Line")
plt.xlabel("Points")
plt.ylabel("Price ($)")
plt.show()
```

---
## Step 6 — Categorical Plots

Analyzing numerical data **grouped by categories**.

```python
# Filter top 5 countries by review count
top_countries = df['country'].value_counts().head(5).index
df_top = df[df['country'].isin(top_countries)]
print("Top 5 Countries:", list(top_countries))
```

```python
# 6.1 — Bar Plot (Average Points by Country)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_top, x='country', y='points', palette='Set2')
plt.title("Average Wine Points by Country")
plt.xlabel("Country")
plt.ylabel("Average Points")
plt.show()
```

```python
# 6.2 — Boxplot (Points Distribution by Country)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top, x='country', y='points', palette='pastel')
plt.title("Wine Points Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Points")
plt.show()
```

```python
# 6.3 — Violin Plot (Distribution + Density)
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_top, x='country', y='points', palette='muted')
plt.title("Wine Points Violin Plot by Country")
plt.xlabel("Country")
plt.ylabel("Points")
plt.show()
```

```python
# 6.4 — Count Plot (Number of Reviews per Country)
plt.figure(figsize=(10, 6))
sns.countplot(data=df_top, x='country', palette='Set1')
plt.title("Number of Wine Reviews by Country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

---
## Step 7 — Heatmap

Heatmaps are the **fastest way to spot correlations** between multiple numerical columns.

```python
# 7.1 — Correlation Heatmap
corr = df[['points', 'price']].dropna().corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap — Points vs Price")
plt.show()
```

---
## Step 8 — Pair Plot

Pair plots give you a **full variable overview** in one single line of code.

```python
# 8.1 — Pair Plot
df_pair = df[['points', 'price']].dropna().sample(500, random_state=42)

sns.pairplot(df_pair)
plt.suptitle("Pair Plot — Points & Price", y=1.02)
plt.show()
```

---
## Step 9 — Seaborn Themes & Styling

```python
# Available Seaborn styles
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

fig, axes = plt.subplots(1, 5, figsize=(18, 4))

for i, style in enumerate(styles):
    with sns.axes_style(style):
        ax = axes[i]
        sns.histplot(df['points'], kde=True, ax=ax, color='steelblue')
        ax.set_title(f"Style: {style}")
        ax.set_xlabel("")

plt.suptitle("Seaborn Style Comparison", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Setting a global style for all subsequent plots
sns.set_style("whitegrid")
sns.set_palette("Set2")

plt.figure(figsize=(10, 5))
sns.histplot(df['points'], kde=True)
plt.title("Whitegrid Style with Set2 Palette")
plt.show()
```

## Step 10 — When to Use What

| Chart | Best Used For | Example |
|-------|---------------|---------|
| `histplot` | Distribution of numerical data | Wine points distribution |
| `kdeplot` | Smooth distribution curve | Visualizing data shape |
| `boxplot` | Outlier detection | Price outliers by country |
| `violinplot` | Distribution + density combined | Points spread per country |
| `scatterplot` | Relationship between 2 numerical columns | Points vs Price |
| `lineplot` | Trend with confidence interval | Avg price per score |
| `regplot` | Scatter + regression trend line | Correlation direction |
| `barplot` | Numerical vs Categorical mean | Avg points by country |
| `countplot` | Frequency of categories | Reviews per country |
| `heatmap` | Correlation between multiple columns | Points & Price correlation |
| `pairplot` | Quick overview of all variable relationships | Full dataset snapshot |

---
## 🔑 Key Takeaways

→ Seaborn is built **on top of Matplotlib** — cleaner syntax, better default styling  
→ `kde=True` in `histplot` adds a smooth distribution curve automatically  
→ `violinplot` shows both **distribution and density** — more informative than a boxplot alone  
→ Heatmaps are the fastest way to **spot correlations** between columns  
→ `pairplot` gives you a full variable overview in **one line of code**  
→ Always use `set_style()` to make your charts **presentation-ready**  
→ Matplotlib gives you **control**. Seaborn gives you **speed**. Know when to use which.
