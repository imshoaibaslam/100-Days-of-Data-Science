## Step 1 — DataFrame from Arrays

```python
data_frame = np.array([[1,2,3],[3,5,6],[8,9,7]])
data_frame
```

```python
new_dataframe =pd.DataFrame(data_frame,index=['row1', 'row2','row3'], columns=['Column No1', 'Column No2', 'Column No3'])
new_dataframe
```

## Step 2 — DataFrame from Dictionary

```python
# list used for this example
states = ['California', 'Texas', 'Florida', 'united Kingdom']
population = [98759, 45888, 787888, 9988884]
```

```python
# Sorting out the list inside the Dictionary
Dict_states = {
    'States': states , 'Population': population
}
```

```python
# Creating the Dataframe now using the same method of pandas
all_population = pd.DataFrame(Dict_states)
```

```python
# Now Showing the Dataframe
all_population
```

## 🔑 Key Takeaways

- A **DataFrame** is a 2D labeled table — think of it as a spreadsheet inside Python
- All three methods produce the **same DataFrame structure** — only the input format differs
- Always inspect your DataFrame right after creating it: `.head()` · `.shape` · `.info()` · `.dtypes`
- **Arrays** → best for numeric / computed data
- **Dictionary** → best for labeled, manual data
- **CSV** → best for external, real-world data
- Missing values load as `NaN` — always check with `df.isnull().sum()`
