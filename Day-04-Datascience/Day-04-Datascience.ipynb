## Step 1 — Summary Functions

Summary functions give us a **quick health report** of our data.
Instead of reading 130,000 rows, we extract key statistics instantly.

#### `describe()` — On a Numeric Column

Think of this as a **one-line report** for any numeric column.

| Stat | Meaning |
|------|---------|
| `count` | How many non-null values exist |
| `mean` | The average value |
| `std` | How spread out the values are |
| `min` | Smallest value |
| `25%` | 25% of values fall below this |
| `50%` | The middle value (median) |
| `75%` | 75% of values fall below this |
| `max` | Largest value |

````python
# Summary of the numeric points column
reviews.points.describe()
````

#### `describe()` — On a Text Column

`describe()` is **type-aware** — it behaves differently for text vs numbers.

For text columns it returns:
- `count` → non-null values
- `unique` → how many distinct values exist
- `top` → the most frequent value
- `freq` → how many times top appears

````python
# Summary of the text taster_name column
reviews.taster_name.describe()
````

#### `mean()` — Average Value

When you only need **one specific statistic**, use focused functions instead of `describe()`.

````python
# What is the average wine score?
reviews.points.mean()
````

#### `unique()` — All Distinct Values

Use this when you want to know **what categories exist** in a column.

````python
# Which tasters exist in this dataset?
reviews.taster_name.unique()
````

#### `value_counts()` — Frequency of Each Value

One of the **most used functions in real data work.**
It tells you each unique value AND how many times it appears.

````python
# Which taster reviewed the most wines?
reviews.taster_name.value_counts()
````

## Step 2 — Mapping with `map()`

### 🗺 What is Mapping?

Mapping means:
> Take each value → apply a rule → produce a new value

Example:
```
Old Value    →    New Value
87           →    -1.44  (below average)
90           →    +1.55  (above average)
```

This is called **remeaning** — centering the data around 0.

### What is `lambda`?

A `lambda` is just a quick one-line function.

Instead of writing:
```python
def subtract_mean(p):
    return p - mean
```
We write:
```python
lambda p: p - mean
```
Same thing. Just shorter.

````python
# Step 1 — Calculate the mean
review_points_mean = reviews.points.mean()
print("Average score:", review_points_mean)
````

````python
# Step 2 — Subtract mean from every value using map()
# For each point value p → return p minus the mean
reviews.points.map(lambda p: p - review_points_mean)
````

**What just happened?**

- `map()` went through every single row in `points`
- For each value, it subtracted the average
- Negative result → that wine scored **below average**
- Positive result → that wine scored **above average**
- `map()` returns a **new Series** — the original data is untouched

## Step 3 — Mapping with `apply()`

### What is `apply()`?

- `map()` works on **one column** (a Series)
- `apply()` works on the **entire DataFrame** — row by row or column by column

### What is a Function?

A function is a **reusable block of code** that takes an input, does something, and returns an output.

```
Input → Function → Output
  87  →  remean  →  -1.44
```

Structure:
```python
def function_name(parameter):
    # do something
    return result
```

### `axis='columns'` vs `axis='index'`

| axis | Meaning | Direction |
|------|---------|----------|
| `axis='columns'` | Apply function **row by row** | → horizontal |
| `axis='index'` | Apply function **column by column** | ↓ vertical |

````python
# Define a function that takes a row and adjusts its points
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

# Apply it row by row across the entire DataFrame
reviews.apply(remean_points, axis='columns')
````

**What just happened?**

- `apply()` passed **one row at a time** into `remean_points()`
- Inside the function, we accessed `row.points` and subtracted the mean
- The modified row was returned
- This repeated for all 129,971 rows

⚠️ **Important:** `apply()` does NOT modify the original DataFrame.
It returns a new transformed copy.

````python
# Proof — original data is unchanged
reviews.head(1)
````

## Step 4 — Vectorized Operations (The Fastest Way)

### What is Vectorization?

Instead of going **value by value**, pandas applies the operation to the **entire column at once**.

No loops. No `map()`. No `apply()`.
Just direct math on the column.

This is **much faster** — especially on large datasets like 130,000 rows.

````python
# Same result as map() — but faster and cleaner
reviews.points - review_points_mean
````

**What pandas does behind the scenes:**

- Left side → entire column (many values)
- Right side → single mean value
- pandas subtracts the mean from **every value automatically**

No function needed. No lambda needed. Just math.

#### Combining Two Columns (Vectorized)

````python
# Combine country and region into one column
reviews.country + " - " + reviews.region_1
````

**Result:** Each row combines like this:
```
Italy + " - " + Etna  →  Italy - Etna
```

If `region_1` is `NaN`, the whole result becomes `NaN` — pandas handles this automatically.

## 💡 When to Use What

| Situation | Use |
|-----------|-----|
| Quick overview of a column | `describe()` |
| Single statistic needed | `mean()`, `min()`, `max()` |
| What categories exist | `unique()` |
| Frequency of each value | `value_counts()` |
| Transform one column value by value | `map()` |
| Complex logic across multiple columns | `apply()` |
| Simple math between columns | Vectorized operators |

---

## 🌍 Real World Usage

**`describe()` + `value_counts()`**
The first two things a data scientist runs after loading any new dataset. They instantly reveal the shape, distribution, and quality of the data.

**`map()`**
Used to create transformed columns — price categories, score buckets, label encoding for machine learning features.

**`apply()`**
Used when logic requires multiple columns together — for example, calculating a value score from points divided by price.

**Vectorized Operations**
The industry standard for production code. Faster, cleaner, and the approach all senior data scientists default to.

---

## 🔑 Key Takeaways

- `describe()` is **type-aware** — behaves differently for numbers vs text
- `value_counts()` is one of the **most used** functions in real data work
- `map()` works on a **single column** only
- `apply()` works on **entire rows or columns** of a DataFrame
- Neither `map()` nor `apply()` modifies the original data — they return **new copies**
- **Vectorized operations** are always faster — prefer them over `map()` and `apply()` when possible
- `axis='columns'` → row by row | `axis='index'` → column by column
